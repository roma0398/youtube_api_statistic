```python
def get_channel_stat():
    url = f"https://www.googleapis.com/youtube/v3/channels?part=statistics&id={channel_id}&key={api_key}"
    json_url = requests.get(url)
    data = json.loads(json_url.text)
    return data
```

```python
def get_videos_data(limit=50):
    url = f"https://www.googleapis.com/youtube/v3/search?key={api_key}&channelId={channel_id}&part=id&order=date"
    url += "&maxResults=" + str(limit)
    vid, npt = get_channel_videos(url)
    idx = 0
    while npt is not None and idx < 4:
        next_url = url + "&pageToken=" + npt
        next_vid, npt = get_channel_videos(next_url)
        vid.update(next_vid)
        idx += 1
    return vid
```

```python
def get_channel_videos(url):
    json_url = requests.get(url)
    data = json.loads(json_url.text)
    channel_videos = dict()
    if "items" not in data:
        return channel_videos, None
    item_data = data["items"]
    next_page = data.get("nextPageToken", None)
    for item in item_data:
        try:
            kind = item["id"]["kind"]
            if kind == "youtube#video":
                video_id = item["id"]["videoId"]
                channel_videos[video_id] = dict()
        except KeyError:
            print("error")
    return channel_videos, next_page
```

```python
def get_video_stat(video_id, part):
    url = f"https://www.googleapis.com/youtube/v3/videos?part={part}&id={video_id}&key={api_key}"
    json_url = requests.get(url)
    data = json.loads(json_url.text)
    try:
        data_part = data["items"][0][part]
    except KeyError:
        print("error")
        data_part = dict()
    return data_part
```

```python
all_video = get_videos_data()
print(all_video)
```

```python
parts = ['snippet', 'statistics', 'contentDetails']
for video_id in tqdm(all_video):
    for part in parts:
        data = get_video_stat(video_id, part)
        all_video[video_id].update(data)
    sleep(1)
print(all_video)
```

```python
a_file = open("data.json", "w")
json.dump(all_video, a_file)
a_file.close()
```

```python
data = pd.read_json('data.json')
data = data.T
data.drop(['channelId', 'thumbnails', 'channelTitle', 'categoryId', 
           'liveBroadcastContent', 'defaultAudioLanguage', 'defaultLanguage', 
           'localized', 'favoriteCount', 'dimension', 'definition', 'caption', 
           'contentRating', 'projection'], axis=1, inplace=True)
```

```python
data['viewCount'] = pd.to_numeric(data['viewCount'])
data['likeCount'] = pd.to_numeric(data['likeCount'])
data['dislikeCount'] = pd.to_numeric(data['dislikeCount'])
data['commentCount'] = pd.to_numeric(data['commentCount'])
```

```python
fig = go.Figure()

for i in ['viewCount', 'likeCount', 'dislikeCount', 'commentCount']:
    fig.add_trace(go.Scatter(mode='lines+markers', x=data['publishedAt'], y=data[i], name=i))
fig.update_traces(marker=dict(size=10, line=dict(width=2,
                                        color='DarkSlateGrey')))
fig.update_layout(legend=dict(x=.9, xanchor="left"),
                  margin=dict(l=0, r=0, t=30, b=0),
                  xaxis=dict(
                  rangeselector=dict(
                  buttons=list([
                        dict(count=1,
                             label="1m",
                             step="month",
                             stepmode="todate"),
                        dict(count=6,
                             label="6m",
                             step="month",
                             stepmode="backward"),
                        dict(count=1,
                             label="YTD",
                             step="year",
                             stepmode="todate"),
                        dict(count=1,
                             label="1y",
                             step="year",
                             stepmode="backward"),
                        dict(step="all")
                  ])),
                  rangeslider=dict(visible=True),type="date"),
                  title="Trend by videos from channel Cut The Crap",
                  title_x=0.5,
                  yaxis_title="Count",
                  xaxis_title="Date"
                 )

fig.show()
fig.write_html("all.html")
```

Метрика для выявления лучших видео, определена как сумма количества просмотров, количество лайков умноженное на 5 и количество комментариев умноженное на 10 и минус колическво дизлайков умноженное на 3, вся сумма делится на 1000 для получения читаемых чисел. Веса для лайков, дизлайков и комментариев были выбраны 

```python
data['metric'] = (data['viewCount'] + data['likeCount'] * 5 + \
                  data['commentCount'] * 20 - data['dislikeCount'] * 3) / 1000
```

```python
top_data = data.sort_values(by=['metric'], ascending=False)[:20].sort_values(by='publishedAt')
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(mode='lines+markers', x=top_data['publishedAt'], y=top_data['metric']))
fig.update_traces(marker=dict(size=10, line=dict(width=2,
                                        color='DarkSlateGrey')))
fig.update_layout(title="Top-20 videos on channel Cut The Crap",
                  title_x=0.5,
                  yaxis_title="metric",
                  xaxis_title="date")
fig.show()
fig.write_html("top-20.html")
```
